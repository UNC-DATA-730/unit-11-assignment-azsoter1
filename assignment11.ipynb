Let's build a random forest model to predict `MedianHouseValue`.

1. Create your model, set the engine to `ranger` with `importance = 'impurity'` and the mode as `regression`.
2. Fit the model using `MedianHouseValue` as the response variable and all other columns as predictors.

````r
mod = rand_forest() |>
    set_engine('ranger', importance = 'impurity',) |>
    set_mode('regression')

mod_fit = mod |> fit(MedianHouseValue ~ ., data = housing)

mod_fit
````

❓ What is the OOB error for this model?

**Answer:**

0.2327352

This is basically the average Residual Sum of Squares (Mean Squared Error). The lower the better (this 0.2327352 value is considered generally low in my view) as it idicates a low difference between the predicted and actual values.

Low OOB prediction error (MSE) indicates high R squared (OOB) value and vice versa.

In this case the R squared (OOB) is 0.8252234 (on a 0-1 scale), which indicates that the model explains a high proportion of the variances in the dependent variable (median house value). For context, if R squared is 0 it would mean that the model does not explain any of the variance in the dependent variable, whereas a model with an R squared value of 1 would be able to explain all the variances in the dependent variable.


Plot the feature importances using the `vip` package. Remember that you can use the `extract_fit_engine` function to get the `ranger` model object to send to the `vip` function.
